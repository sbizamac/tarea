```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

# Lectura y Formateo de data

## Tratamiento de datos:
 - Lectura de CSV
 - Encodeo de formato
 - Separación de Columnas por punto y coma
 - Creación de nombres de columna, para la correcta interpretación de data y eliminación de errores en formato, puntualmente en la columna Fecha
 - Además creamos una nueva columna con la fecha formateada a "DateTime", para un correcto tratamiento de las semanas

```python
df = pd.read_csv('fallecidos_covid.csv', encoding = "ISO-8859-1",sep=";")
df.head()
```

> Vemos problemas de título extraño en el campo fecha_fallecimiento

```python
df.columns = ['Fecha', 'Region', 'Hospitalizacion', 'Fallecidos']
df['Fecha'] = pd.to_datetime(df['Fecha'])
df.head()
```

> Validamos la correcta creación del campo Fecha, además de su transformación en DateTime a la variable 

```python
## Iniciamos las semanas desde el 00 en adelante. 
df['Week_Number'] = df['Fecha'].dt.strftime('%U')
```

```python
df
```

```python
## Comprobamos la cantidad de datos x semana, vemos que en la semana 00 hay 30 datos, y en la semana 49, como última de la muestra, hay 90
df.Week_Number.value_counts()
```

```python
df
```

```python
# Ordenamos desde la semana 00 a la semana 49 ascendentemente
df.sort_values(by=['Week_Number'],inplace=True)
```

```python
df
```

```python
## Graficamos Q de fallecidos x Semana. Podemos deducir que el Peak de fallecidos se concentra alrededor de la semana 22.
df.plot.line(x='Week_Number', y='Fallecidos')
```

```python
col_list = ["2021-03-24", "Tumores malignos de los órganos digestivos"]
df = pd.read_csv("DEFUNCIONES_FUENTE_DEIS_2016_2021_29042021.csv", encoding = "ISO-8859-1",sep=";",usecols=col_list)
```

```python
df
```

```python
df.columns = ['Fecha', 'Fallecimiento_No_COVID']
df_refine= df[df.Fecha.str.contains('2020')] ## DATAFRAME SOLO CON FECHAS 2020
df_refine.head()
```

```python
df_refine
```

```python
df_influenza=df_refine[df_refine.Fallecimiento_No_COVID.str.startswith('Influenza')] 
```

```python
df_influenza
```

Se procede a identificar Influenza en la muestra de enfermedades causales:

```python
df_influenza.value_counts()
```

```python
df_influenza['Fecha_2'] = pd.to_datetime(df_influenza['Fecha'])
```

```python
df_influenza['Week_Number'] = df_influenza['Fecha_2'].dt.strftime('%U')
```

```python
df_influenza.head()
```

```python
df_influenza.sort_values(by=['Week_Number'],inplace=True)
```

```python
df_influenza.Week_Number.value_counts()
```

```python
df_influenza['Fallecimiento_No_COVID'] = np.where( (df_influenza.Fallecimiento_No_COVID == 'Influenza [gripe] y neumonía') , 1, 0)  # Marca 0 si la causa fue COVID, marca 1 si la causa no fue 
```

```python
df_influenza.plot.line(x='Week_Number', y='Fallecimiento_No_COVID')
```
